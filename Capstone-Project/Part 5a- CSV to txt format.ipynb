## Converting reviews to txt format

- This part of the notebook is used to convert positive reviews from csv format to txt format.
- The txt format data is then used inside colab.
- The reviews are chosen from `course_id` python course as it is the most reviewed course.
- Reviews with rating = 5 are chosen for text-generation as the aim is to produce positive reviews
- Also, post with more than 30 words are chosen, this is because reviews with low diversity in text will produce repeated words.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('trans_df.csv')
```

```python
df['course_id'].value_counts().head(10)
```

```python
python_rating_5 = df[(df['course_id'] == 'python')&(df['rating']==5)]
```

```python
python_rating_5.head(2)
```

```python
python_rating_5.shape
```

```python
python_rating_5[python_rating_5['count of words']>30]
```

```python
rating_5_python_minimum_30= python_rating_5[python_rating_5['count of words']>30]
```

```python
python_rating_5['reviews'].values
```

```python
rating_5_python_minimum_30['reviews'].to_csv("rnn_text.txt", header=False, index=False)
```
